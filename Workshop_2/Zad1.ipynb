```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
from sklearn.linear_model import LinearRegression

```

## EDA


```python
path = 'C:/Users/matys/Downloads/Pokemons/'
file_name = 'pokedex_(Update_05.20).csv'
df = pd.read_csv(path + file_name, sep=',')

df = df.drop(['Unnamed: 0', 'german_name', 'japanese_name'], axis=1)
df.head(10)
```

```python
how_many_nulls = df.isnull().sum()

to_delete = []

for index, number in enumerate(how_many_nulls):
    if number > 150:
        to_delete += [index]

df = df.drop(df.columns[to_delete],axis=1)
df = df.dropna()
df.isnull().sum().sum()
```

```python
df.head(10)
```

## Preprocessing data

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

Y = df['total_points']
X = df.drop(['name', 'total_points', 'pokedex_number', 'type_number', 'abilities_number', 'egg_type_number', 'species'], axis=1)

numerical = X.select_dtypes(include=np.number).drop(['generation'], axis=1)
len(numerical.columns)

column_transformer = ColumnTransformer([('categorical', OneHotEncoder(drop='first'),
 ['status', 'type_1', 'ability_1','growth_rate', 'egg_type_1', 'generation']),
('numerical', StandardScaler(), numerical.columns)])

X = column_transformer.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.75)
```

## Pipeline and training model

```python

pipeline = Pipeline(steps=[ ('feature_selection', SelectFromModel(LinearRegression())),
                            ('clf', LinearRegression())                                            
])

pipeline.fit(x_train, y_train)
preds = pipeline.predict(x_test)
```

## Results

```python
results = [f'{list(y_test)[i]}, {list(preds)[i]}' for i in range(len(list(y_test)))]
results[:10]
```

```python
squared_error = sum([(list(y_test)[i] - list(preds)[i]) ** 2 for i in range(len(list(y_test)))])
squared_error
```
